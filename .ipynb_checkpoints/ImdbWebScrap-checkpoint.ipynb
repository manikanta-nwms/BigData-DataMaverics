```python
import requests
from bs4 import BeautifulSoup
import csv
import re


base_url = 'https://www.imdb.com'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}


csv_file = open('top_250movies_96_22.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Ranking', 'IMDByear', 'IMDBlink', 'Title', 'RunTime', 'Genre', 'Rating', 'Score', 'Votes', 'Gross', 'Director', 'Cast1', 'Cast2', 'Cast3', 'Cast4'])


for year in range(1996, 2023):
    print(f'Scraping top 250 movies of {year}...')
    for start in range(1, 251, 50):
        
        url = f'{base_url}/search/title/?title_type=feature&release_date={year}-01-01,{year}-12-31&user_rating=1.0,10.0&num_votes=5000,&sort=user_rating,desc&start={start}&ref_=adv_nxt'

    
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')

      
        movie_items = soup.find_all('div', class_='lister-item-content')

       
        for i, item in enumerate(movie_items):
            ranking = start + i
            title_link = item.h3.a
            title = title_link.text
            imdb_link = base_url + title_link['href']
            year = int(re.search('\d{4}', item.find('span', class_='lister-item-year').text).group())
            runtime = item.find('span', class_='runtime').text.strip(' min')
            genre = item.find('span', class_='genre').text.strip()
            rating = float(item.strong.text)
            score = float(item.find('span', class_='metascore').text) if item.find('span', class_='metascore') else None
            votes = int(item.find('span', attrs={'name': 'nv'})['data-value'])
            gross_items = item.find_all('span', attrs={'name': 'nv'})
            gross = gross_items[1].text.strip('$') if len(gross_items) > 1 else None
            #gross = item.find_all('span', attrs={'name': 'nv'})[1].text.strip('$') if len(item.find_all('span', attrs={'name': 'nv'})) > 1 else None
            director = item.find_all('p')[2].a.text
            casts = [cast.text for cast in item.find_all('p')[2].find_all('a')[1:5]]
           
           
        
            # Write the data to the CSV file
            csv_writer.writerow([ranking, year, imdb_link, title, runtime, genre, rating, score, votes, gross, director, *casts])
        
```

## Cleaning data manipulating data. having some empty data rows need to delete

```python
import pandas as pd


df = pd.read_csv('top_250movies_96_22.csv', delimiter=',')

# Renaming the columns for better readability
df = df.rename(columns={'Ranking': 'Rank', 'IMDByear': 'Year', 'IMDBlink': 'Link', 'RunTime': 'Duration','Genre':'Genre',
                        'Cast1': 'Actor1', 'Cast2': 'Actor2', 'Cast3': 'Actor3',
                        'Cast4': 'Actor4'})

#remove score 
df = df.drop(['Score'], axis=1)



# Splitting the Genre 
df[['Genre1', 'Genre2', 'Genre3']] = df['Genre'].str.split(',', expand=True)

# Reordering the columns for better readability
df = df[['Rank', 'Title', 'Year', 'Duration', 'Genre1', 'Genre2', 'Genre3', 'Rating', 'Votes', 'Director', 'Actor1', 'Actor2', 'Actor3', 'Actor4', 'Link']]

# Saving the manipulated data as 
df.to_csv('manipulated_data96-22.csv', index=False)


```

**Q: Is there a significant change in the average rating of top 250 IMDb movies over the years??**

A: produce a line graph that shows the trend of the average rating for the top 250 movies on IMDb from 1996 to 2022.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cleaned_data96-22.csv')

# Calculate the mean rating for each year
mean_ratings = df.groupby('Year')['Rating'].mean()


plt.plot(mean_ratings.index, mean_ratings.values)
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Average Rating of Top 250 IMDb Movies (1996-2022)')
plt.show()

```


<font color='blue'>**Q:** What are the most common genres among the top 250 IMDb movies from 1996 to 2022?</font>

<font color='green'>**A:** produce a bar chart that shows the frequency of the top 10 most common genres in the top 250 IMDb movies from 1996 to 2022.</font>

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('cleaned_data96-22.csv')





df = df.dropna()

# Split the genres and count their frequency
genres = df['Genre1'].str.cat(df[['Genre2', 'Genre3']], sep=',').str.split(',')
genre_counts = pd.Series([genre.strip() for sublist in genres for genre in sublist]).value_counts().nlargest(10)

# bar chart
plt.bar(genre_counts.index, genre_counts.values)
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.title('Top 10 Most Frequent Genres in Top 250 IMDb Movies (1996-2022)')
plt.xticks(rotation=90)
plt.show()

```
